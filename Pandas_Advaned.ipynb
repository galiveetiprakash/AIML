ExcelFile Class : To facilitate working with multiple sheets from the same file, the ExcelFile class can be used to wrap the file and can be passed into read_excel There will be a performance benefit for reading multiple sheets as the file is read into memory only once.

```
#Write a Code to Connect Excel sheet
import pandas as pd
Exceldata = pd.ExcelFile('forestfire.xlsx')
Forestdata = pd.read_excel(Exceldata, 'phase1')
Forestdata
```

```
#Write a Code to load the data from Excel sheets (2 Sheets)
with pd.ExcelFile("forestfire.xlsx") as xls:
  phase1=pd.read_excel(xls,"phase1")
  phase2=pd.read_excel(xls,"phase2")

phase1.shape
phase2.shape
```

Selection By Position :Pandas provides a suite of methods in order to get purely integer based indexing.

Merge, Join and Concat

```
#Concatenating Objets
data_2012 = phase1[(phase1['year']==2012)]
data_2013 = phase1[(phase1['year']==2013)]

data_2012_2013=[data_2012,data_2013]
pd.concat(data_2012_2013)
```

```
# Concat all the records from Phase1 and Phase2
phase1.count()
phase2.count()
#Using keys we can add Hierarchical Index at the outmost level
totals=[phase1,phase2]
Forestdata=pd.concat(totals,keys=['p1','p2'])
Forestdata.loc['p1'].count()
Forestdata.loc['p2'].count()
```

```
# clear the existing Index and reset it in the result by setting the ignore_index option to True.
Forestdata=pd.concat(totals,ignore_index=True)
Forestdata
```

```
#Concating two dataframes with different structures
skills = pd.DataFrame({'skill': ['oracle', 'db2', 'java', 'python'], 'demand': ['high', 'low', 'low', 'high'],
'cost': [100, 80, 20, 200]})

location = pd.DataFrame({'region': ['Hyderabad', 'bangaore', 'Chennai', 'Pune'], 'Experience': [4,6,7,8]})

Rating =pd.DataFrame({'Type': ['Happy', 'VeryHappy', 'Average', 'Poor'], 'Experience': [1,2,3,4]})
Alldata = pd.concat([skills,location],axis=1)
Alldata
```

```
#Concatenating Using Append()

Allphases = phase1.append(phase2)
Allphases

#Append helps to concatenate multiple Objects
Alldat= skills.append([location,Rating],sort=True)
Alldat
```

```
#Joins
skills = pd.DataFrame({'skill': ['oracle', 'db2', 'java', 'python'], 'demand': ['high', 'low', 'low', 'high'],
'cost': [100, 80, 20, 200]},index=['A0','A1','A2','A4'])

location = pd.DataFrame({'region': ['Hyderabad', 'bangaore', 'Chennai', 'Pune'], 'Experience': [4,6,7,8]},index=['B1','B2','B3','B4'])

Rating =pd.DataFrame({'Type': ['Happy', 'VeryHappy', 'Average', 'Poor'], 'ID': [1,2,3,4]},index=['A0','A1','A3','A5'])

print(skills)
print(location)
print(Rating)

alldata =skills.join(location)
print(alldata)

alldata =skills.join(location,how='outer')
print(alldata)

alldata =skills.join(Rating,how='inner')
print(alldata)
```
